## Teksta lejuplāde un priekšapstrāde. Mainīgajā 'url' jāieraksta tīmekļa lapas saite

```python
import requests
from bs4 import BeautifulSoup

# Function to crawl a webpage and extract all links
def extract_text_from_webpage(url):
    try:
        # Send a GET request to the webpage
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the page content with BeautifulSoup
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract and print the text from the webpage
            page_text = soup.get_text()
            print(page_text)

            # Clean up the text by removing excessive whitespace
#            clean_text = ' '.join(page_text.split())
#            print(clean_text)
            return(page_text)
        else:
            print(f"Failed to access {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Mainīgajā 'url' jāieraksta lejuplādējamās lapas adrese
# šo lapu varēs lejuplādēt

url = 'https://korpuss.lv/klasika/Senie/RBl/RBl0122.htm'

# šo lapu nevarēs lejuplādēt sarežģītās lapas struktūras dēļ
# url=''https://satori.lv/article/velreiz-par-ergliem-un-cuskam-kuru-nav


# šo lapu nevarēs lejuplādēt, jo aizsargāta pret kopēšanu
#url='https://leta.lv/home/important/92ADB76A-804E-4870-9652-2B641C77D277/'
mytext = extract_text_from_webpage(url)
```

Teksta saglabāšana failā

```python
# mainīgajā filename ierakstām faila nosaukumu
filename = "velnini.txt"

with open(filename, 'w') as file:
      file.write(mytext)

```

## Cikliska tīmekļa lapu vākšana


```python
import requests
from urllib.parse import urljoin, urlparse


# iegūst saišu/lapu sarakstu no galvenās lapas
def get_all_links(url):
    """Fetch all valid subpage links from a webpage."""
    try:
        content = requests.get(url).content
        soup = BeautifulSoup(content, 'html.parser')
        links = []
        for link in soup.find_all('a'):
         if link.has_attr('href'):
            href = link['href']
            if href and not href.startswith('#'):
                links.append(sakums+href)
        return links


    except Exception as e:
        print(f"Error fetching links from {url}: {e}")
        return set()

# iegūst lapas
def scrape_website_with_subpages(main_url):
    """Scrape text from the main page and all its subpages."""
    all_text = ""

    # Get text from the main page
    # print(f"Scraping main page: {main_url}")
    # all_text += get_text_from_url(main_url)

    # Find all subpage links
    subpages = get_all_links(main_url)

    # Scrape text from each subpage
    for subpage_url in subpages:
        print(f"Scraping subpage: {subpage_url}")
#        all_text += get_text_from_url(subpage_url)
        all_text += '\n' + extract_text_from_webpage(subpage_url)

    return all_text

# Example usage
main_website_url = "https://korpuss.lv/klasika/Senie/RBl/RBl01.htm"
sakums = "https://korpuss.lv/klasika/Senie/RBl/"
all_website_text = scrape_website_with_subpages(main_website_url)
print(all_website_text)

# mainīgajā filename ierakstām faila nosaukumu
filename = "blaumanis.txt"

with open(filename, 'w') as file:
      file.write(all_website_text)
```

Vēl viens veids kā izgūt tekstu


```python
# iegūst lapas saturu un tekstu
def get_text_from_url(url):
    """Fetch and parse text from a single webpage."""
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        # Get all the text from the page (remove scripts and styles)
        for script_or_style in soup(["script", "style"]):
            script_or_style.extract()

        return soup.get_text(separator=' ', strip=True)
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return ""
```
